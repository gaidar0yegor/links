```python
import requests
from bs4 import BeautifulSoup

url = "https://www.amazon.it/dp/B0FGMNGVDT?th=1&psc=1"
headers = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36'
}

response = requests.get(url, headers=headers)
soup = BeautifulSoup(response.text, 'html.parser')

# –ü—Ä–æ–≤–µ—Ä—è–µ–º —Å–µ–ª–µ–∫—Ç–æ—Ä—ã —Ä–µ–π—Ç–∏–Ω–≥–∞
selectors = [
    '#averageCustomerReviews .a-icon-star .a-icon-alt',
    '#acrPopover .a-icon-alt',
    '.reviewCountTextLinkedHistogram .a-icon-alt',
    '#acrCustomerReviewText',
]

for selector in selectors:
    element = soup.select_one(selector)
    if element:
        print(f"‚úÖ {selector}: '{element.get_text().strip()}'")
    else:
        print(f"‚ùå {selector}: NOT FOUND")

# –¢–∞–∫–∂–µ –≤—ã–≤–µ–¥–µ–º –≤—Å–µ —ç–ª–µ–º–µ–Ω—Ç—ã —Å "stelle" –≤ —Ç–µ–∫—Å—Ç–µ
for el in soup.find_all(string=lambda t: t and 'stelle' in t.lower()):
    print(f"üîç Found 'stelle': {el.strip()[:100]}")
```
